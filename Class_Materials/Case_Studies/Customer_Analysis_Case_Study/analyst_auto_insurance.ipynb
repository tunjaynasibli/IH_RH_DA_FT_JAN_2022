```python
import pandas as pd
```

```python
data = pd.read_csv("Data/file1.csv")
```

```python
columns_lower = []
data.columns
def standardize(df):
    for column in df.columns:
        columns_lower.append(column.lower().replace(" ", "_"))
    df.columns = columns_lower 

    return df 
standardize(data)
```

#### Deleting and rearranging columns – delete the column customer as it is only a unique identifier for each row of data


```python
data.drop("customer", axis=1, inplace=True)
data
```

```python
clv = data["customer_lifetime_value"]
def clean_data(x):
    for i in x:
        if i[-1] == "%":
            x.rstrip
        return x
clean_data(clv)    
```
